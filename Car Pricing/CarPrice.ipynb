```python
numeric_feat = dataset.describe().columns
numeric_feat
```

```python
#plot bar graph for every feature expect (car_id)
for col in numeric_feat[2:]:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = dataset[col]
    feature.hist(bins=50, ax = ax)
    ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
    ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)
    ax.set_title(col)
plt.show()
```

```python
#log transformation
for col in numeric_feat[2:]:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = np.log(dataset[col])
    feature.hist(bins=50, ax = ax)
    ax.axvline(feature.mean(), color='magenta', linestyle='dashed', linewidth=2)
    ax.axvline(feature.median(), color='cyan', linestyle='dashed', linewidth=2)
    ax.set_title(col)
plt.show()
```

```python
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x'], point['y'], str(point['val']))
```

```python
for col in numeric_feat[1:-1]:
    fig = plt.figure(figsize=(9, 6))
    ax = fig.gca()
    feature = dataset[col]
    label = dataset['price']
    
    # Calculate correlation
    correlation = feature.corr(label)
    
    # Scatter plot
    plt.scatter(x=feature, y=label)
    plt.xlabel(col)
    plt.ylabel("Price")
    
    # Set title with correlation value
    ax.set_title('price vs ' + col + '- correlation: ' + str(correlation))
    
    # Linear fit (1st-degree polynomial)
    z = np.polyfit(feature, label, 1)
    y_hat = np.poly1d(z)(dataset[col])  # Use np.poly1d to create a polynomial
    
    # Plot trendline
    plt.plot(feature, y_hat, "r--", lw=1)

# Show all plots
plt.show()
```
